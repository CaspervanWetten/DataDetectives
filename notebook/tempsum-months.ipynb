```python
import pandas as pd
import numpy as np
```

```python
tempdf = pd.read_csv("TemperatureData.csv", delimiter=",")
tempsel = tempdf.loc[tempdf["Date"] >= "1986-01-01"]
tempsel["Date"] = tempsel["Date"].str[:-3]
```

```python

#tempsel["Date"] = pd.to_numeric(tempsel["Date"])
dates = pd.unique(tempsel["Date"])
#tempsel.loc[tempsel["Date"] == dates[300], tempsel.columns != "Date"].mean()
```

```python
meanlist = []
for date in dates:
    a = tempsel.loc[tempsel["Date"] == date, tempsel.columns != "Date"].mean()
    meanlist.append(a)
tempmeans = pd.DataFrame(meanlist)
```

```python
tempdf2 = tempmeans.T
tempdf2.columns = dates
tempdf2.head()
```

```python
tempdf.columns = dates
tempdf = tempdf.drop(tempdf.index[0])
tempdf2 = tempdf.reset_index()
tempdf2.rename(columns = {'index':'Country_Code'}, inplace = True)
tempdf2.head()

```

```python
tempdf2.to_csv("tempmeans.csv")
```

```python
import pandas as pd
```

```python
tempdf = pd.read_csv("TemperatureData.csv", delimiter=",")
tempsel = tempdf.loc[tempdf["Date"] >= "1986-01-01"]
tempsel["Date"] = tempsel["Date"].str[:-6]
```

```python
tempsel.head()
tempsel["Date"] = pd.to_numeric(tempsel["Date"])
dates = tempsel["Date"].unique()
```

```python
meanlist = []
for date in dates:
    a = tempsel.loc[tempsel["Date"] == date].mean()
    meanlist.append(a)
tempmeans = pd.DataFrame(meanlist)
tempmeans['Date'] = tempmeans['Date'].astype(str)
tempmeans["Date"] = tempmeans["Date"].str[0:4]
```

```python
tempmeans.head()
tempdf = tempmeans.T
```

```python
tempdf.columns = tempdf.iloc[0]
tempdf = tempdf.drop(tempdf.index[0])
tempdf2 = tempdf.reset_index()
tempdf2.rename(columns = {'index':'Country_Code'}, inplace = True)
tempdf2.head()

```

```python
tempdf2.to_csv("tempmeans.csv")
```
